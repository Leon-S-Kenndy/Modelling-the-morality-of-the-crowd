```python
def clean_scrape(df):
  # Remvoe some comments that are too short
  
  # Replace some common alternate spellings
  df['text'] = df['text'].replace(r'\n',' ', regex=True) 
  df['text'] = df['text'].str.replace("YTA","You're the Asshole")
  df['text'] = df['text'].str.replace("NTA","You're Not the Asshole")
  df['text'] = df['text'].str.replace("ESH","Everyone Sucks Here")
  df['text'] = df['text'].str.replace("NAH","No Asshholes here")
  df['text'] = df['text'].str.replace("INFO","Not Enough Info")
  # Remove any deleted or removed posts
  gone_list = ["[deleted]","[removed]"]
  
  df_use = df[df['text'].isin(gone_list)==False]
  print("After removing deleted posts, there are " +  str(len(df_use)) + " posts left.")
#to keep 8
  # Write to file
  #df_use.to_csv("AITA_cleaned.csv")
  return(df_use)
```

```python
data_list = []
dataset = None
for fileitem in files:
  tmp = pd.read_csv(fileitem)
  tmp = tmp[tmp.text.str.count('\s+').ge(10)]
  tmp = clean_scrape(tmp)
  tmp = tmp.dropna()
  data_list.append(tmp)
dataset = pd.concat(data_list,ignore_index = True)
```

```python
grand = clean_scrape(dataset)
print("There are now " +  str(len(grand)) + " cleaned posts.")

grand.to_csv("aita_clean.csv",index=False)
```
